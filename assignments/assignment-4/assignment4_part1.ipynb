```python
import os
import json
from openai import OpenAI
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

OPENAI_API_KEY = "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"

client = OpenAI(api_key=OPENAI_API_KEY)

TRANSCRIPT_FILE = "transcript.srt"

def load_srt(path):
    entries = []
    with open(path, "r") as f:
        block = []
        for line in f:
            line = line.strip()
            if line == "":
                if block:
                    entries.append(block)
                    block = []
            else:
                block.append(line)
        if block:
            entries.append(block)

    chunks = []
    for block in entries:
        if len(block) >= 3:
            times = block[1]
            text = " ".join(block[2:])
            chunks.append({"timestamp": times, "text": text})
    return chunks

def build_index(chunks):
    model = SentenceTransformer("all-MiniLM-L6-v2")
    texts = [c["text"] for c in chunks]
    embeddings = model.encode(texts, convert_to_numpy=True)
    dim = embeddings.shape[1]
    index = faiss.IndexFlatL2(dim)
    index.add(embeddings)
    return index, embeddings

def retrieve(query, chunks, index, embeddings, top_k=15):
    model = SentenceTransformer("all-MiniLM-L6-v2")
    q_emb = model.encode([query], convert_to_numpy=True)
    scores, idx = index.search(q_emb, top_k)
    results = []
    for i in idx[0]:
        results.append(chunks[i])
    return results

def answer_query(query, context_chunks):
    context_text = ""
    for c in context_chunks:
        context_text += f"[{c['timestamp']}] {c['text']}\n"

    prompt = f"""
You are an AI assistant analyzing a basketball commentary transcript.

Use only the text provided in CONTEXT to answer the question.
Cite timestamps when useful.

CONTEXT:
{context_text}

QUESTION: {query}

Answer clearly and concisely.
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content

def main():
    print("Loading transcript...")
    chunks = load_srt(TRANSCRIPT_FILE)
    print("Building search index...")
    index, embeddings = build_index(chunks)

    print("Game Q&A system ready.")
    print("Ask a question (Ctrl+C to exit).\n")

    while True:
        query = input("Q: ")
        results = retrieve(query, chunks, index, embeddings)
        answer = answer_query(query, results)
        print("\nA:", answer, "\n")

if __name__ == "__main__":
    main()
```
