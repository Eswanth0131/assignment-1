Mixture of Gaussians Dataset

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.default_rng()

n = 1000
k = 3

means = rng.uniform(-5, 5, size=(k, 2))
covs = [np.diag(rng.uniform(0.3, 1.2, size=2)**2) for _ in range(k)]
weights = rng.dirichlet(np.ones(k))

assignments = rng.choice(k, size=n, p=weights)

X = np.vstack([
    rng.multivariate_normal(means[i], covs[i], size=(assignments == i).sum())
    for i in range(k)
])

plt.scatter(X[:,0], X[:,1], c=assignments)
plt.title('2D Dataset')
plt.show()
```
